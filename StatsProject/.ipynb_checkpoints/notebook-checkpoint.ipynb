```python
from data import *
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
%config InlineBackend.figure_format = 'svg'
```

```python
def func(x, a, b):
    return (x * a) + b
```

```python
def plot(x_data, y_data, color):
    i = [1, 1]
    t = curve_fit(func, x_data, y_data, i)
    residuals = np.array(y_data) - (func(np.array(x_data), *t[0]))
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((np.array(y_data)-np.mean(np.array(y_data)))**2)
    r = (1 - (ss_res / ss_tot))*100
    r_squared = round(r, 3)
    xd = np.linspace(min(x_data), max(x_data), 100)
    line = [func(x, *t[0]) for x in xd]
    rounded = list(map(lambda x: round(x, 3), t[0]))
    l = "{}x+ {} R^2 = {}%".format(*rounded, r_squared)
    plt.scatter(x_data, y_data, color=color)
    plt.plot(xd, line, color=color, label=l)
```

```python
plt.title("Number of People vs Internet Speed")
plt.xlabel(r"Number of People")
plt.ylabel("Internet Speed (Mbps)")
plot(x, d2, color="y")
plt.legend(loc="upper left")
plt.show()
```
